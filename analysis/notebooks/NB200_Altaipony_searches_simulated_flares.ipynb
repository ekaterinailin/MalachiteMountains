```python
from altaipony.altai import equivalent_duration
from altaipony.flarelc import FlareLightCurve
import os
CWD = "/".join(os.getcwd().split("/")[:-2])
import pandas as pd
import numpy as np

from astropy.constants import R_sun, M_sun, G, k_B, m_p
import astropy.units as u

# Plot stuff
from funcs.model import full_model_coupled, aflare

import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.lines import Line2D

import matplotlib 
matplotlib.rc('xtick', labelsize=14) 
matplotlib.rc('ytick', labelsize=14) 

font = {'family' : 'courier',
        'weight' : 'normal',
        'size'   : 14}

matplotlib.rc('font', **font)

import time
metatstamp = time.strftime("%d_%m_%Y", time.localtime())
```

```python
flares = pd.DataFrame()
Prot = .4
time = np.linspace(1.,43.,3200)
phi = (time - 10.) / Prot * np.pi * 2.
qlum = 1e31 *u.erg / u.s
Fth = 90616356361.5605 * u.erg / (u.cm**2 * u.s)
R = .15 * R_sun
median = 300.
phi_a=10.
fwhm = 4.
phi0= 3.4
#plt.figure(figsize=(8,5))
for i_ in [5., 10., 30, 45., 60., 85.]:# [10., 30, 45., 60., 85.]
    for a in [0.25,0.5,1,2.]:#[0.5,1,1.5,2.]
        for theta_a in [5., 10., 30, 45., 60., 85.]: #np.arange(.1,89.9,5)
            i = i_ / 180. * np.pi 
            th = theta_a / 180. * np.pi

            fullmodel = full_model_coupled(phi_a, th, a, fwhm, i, phi0=phi0,
                               phi=phi, num_pts=50, qlum=qlum,
                               Fth=Fth, R=R, median=median)
            fullmodel = fullmodel + fullmodel*np.random.normal(0.00,0.02,len(fullmodel))
            flc = FlareLightCurve(time=time, detrended_flux=fullmodel, detrended_flux_err=.02*fullmodel)
            flc.it_med = np.full_like(time, median)
        #     ED = equivalent_duration(flc,0,-2)
        #     sim[theta_a] = ED * qlum.value
            flar = flc.find_flares().flares
            if len(flar) == 0:
                print("noflares")
                flar = pd.DataFrame(dict(zip(['istart', 'istop', 'cstart', 
                                     'cstop', 'tstart', 'tstop', 
                                     'ed_rec', 'ed_rec_err', 'ampl_rec', 
                                     'dur', 'total_n_valid_data_points', 
                                     'theta', "a_input", "inlc"],
                                    [[np.nan], [np.nan], [np.nan],
                                     [np.nan], [np.nan], [np.nan],
                                     [np.nan], [np.nan], [np.nan],
                                     [np.nan], [len(time)], [theta_a], [a], [i_]])))
            else:
                flar["theta"] = theta_a
                flar["a_input"] = a
                flar["inlc"] = i_
            flares = flares.append(flar)
            
#             plt.plot(flc.time, fullmodel)
```

```python
f = flares.groupby(["theta", "a_input", "inlc"]).apply(lambda x: x.sort_values(by="dur", ascending=False).iloc[0])
```
