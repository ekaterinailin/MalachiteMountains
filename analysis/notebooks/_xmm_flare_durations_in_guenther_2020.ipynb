```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from matplotlib.lines import Line2D
```

```python
df = pd.read_csv("../../data/guenther/guenther_complete.tsv", skiprows=76, delimiter="\t", na_values={'     ':np.nan})
```

```python
df.columns
```

```python
df = df.replace('     ',9999)
df = df.replace('         ', np.nan)
df = df.replace('      ', np.nan)
```

```python
df["FWHMdt"].astype(float).hist()
```

```python
df.SpT.unique()
```

```python
df = df[(~df.Ebol.isnull()) & (~df.aFFD.isnull())]
df["log_freq"] = np.log10(df.Ebol.astype(float).values)*df.aFFD.astype(float).values + df.bFFD.astype(float).values
df["freq"] = np.power(10,df["log_freq"])
```

```python
plt.figure(figsize=(10,7))
legend_elements = []
m = ["X","o","d"]
c = ['c',"r","k"]
for spt in ["M4-M10",  'M0-M4 ', 'FGK   ']:
    d = df[(df.SpT ==spt)]
#     print(d.shape)
    ma, co = m.pop(), c.pop()
    plt.scatter(d["FWHMdt"].astype(float)*24, 
                d["freq"].astype(float), 
                marker=ma, s=25, label=spt, color=co,
               alpha=1)
    

    legend_elements.append(Line2D([0], [0], marker=ma, color="w", label=spt,
                          markerfacecolor=co, markersize=10))
plt.legend(handles=legend_elements, fontsize=16, frameon=False, loc=3)
# plt.legend(loc=3, fontsize=16, frameon=False)
plt.yscale("log")
plt.xscale("log")
plt.xlim(None,2.5)
plt.xlabel("$FWHM$ [h]", fontsize=16)
# plt.ylabel(r"$E_{bol}$ [erg]", fontsize=16)
plt.ylabel(r"flare frequency $>E_{bol}$ [d$^{-1}$]", fontsize=16)
plt.tick_params(labelsize=16)


plt.savefig("/home/ekaterina/Documents/002_writing/xmm-proceedings/figures/guenther_repr_fwhm_freq.png",dpi=300);
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

def colorFader(c1,c2,mix=0): #fade (linear interpolate) from color c1 (at mix=0) to c2 (mix=1)
    c1=np.array(mpl.colors.to_rgb(c1))
    c2=np.array(mpl.colors.to_rgb(c2))
    return mpl.colors.to_hex((1-mix**2)*c1 + mix**3*c2)

c1='white' #blue
c2='black' #green
n=500
for x in np.linspace(0.5,2.5,n):
    plt.axvline(x, color=colorFader(c1,c2,x/6), linewidth=4) 
plt.show()
```
